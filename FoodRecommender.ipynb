```python
for k in range(len(d)):
    for i in range(len(d[k])-1,-1,-1):
        for j in ingredients.keys():
            for x in ingredients[j]:
                if x in d[k][i]:
                    d[k][i]=j
        if d[k][i] not in ingredients.keys():
            del d[k][i]
```

```python
for i in range(d.shape[0]):
    data['مواد اولیه'][i]=d[i]
```

```python
icm=data.copy()
for index, row in data.iterrows():
    for ingred in row['مواد اولیه']:
        icm.at[index, ingred] = 1
icm=icm.fillna(0)
icm.drop(['نام غذا','مواد اولیه'],axis=1,inplace=True)
icm.rename(columns={'کالری برای یک نفر':'calorie',
                   'وعده غذایی':'meal',
                   'کشور':'region'},inplace=True)
```

```python
icm.head(2)
```

```python
icm=pd.get_dummies(icm,columns=['region','meal'] , drop_first=False)
```

```python
scaler = MinMaxScaler()
icm['calorie']=scaler.fit_transform(icm[['calorie']])
```

## testing a user

```python
random_user_1=[
{'id':6,'point':1},
{'id':5,'point':1},
{'id':151,'point':1},
{'id':3,'point':-1},
{'id':15,'point':-1},
{'id':17,'point':-1},
]
random_user_1= pd.DataFrame(random_user_1)
```

```python
pd.merge(data,random_user_1, how='inner', left_on = 'ID', right_on = 'id').drop('id',axis=1)
```

```python
userfeature=icm[icm['ID'].isin(random_user_1['id'].tolist())]
userfeature=userfeature.reset_index(drop=True).drop('ID',1)
```

```python
userfeature
```

```python
userProfile=userfeature.transpose().dot(random_user_1['point'])
```

```python
pd.DataFrame(userProfile)
```

```python
recommendation=(((icm.drop(icm[icm['ID'].isin(random_user_1['id'].tolist())].index).set_index('ID')*userProfile).sum(1)/(userProfile.sum())).sort_values(ascending=False)).reset_index(drop=False)
```
